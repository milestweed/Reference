# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#This code enables inline display of matplot lib plots
#Otherwise use plt.show()
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### Functional method

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r*')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

#[left,bottom,width,height]
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes #axes is a list of axes objects
```

```python
# can be iterated over!
fig,axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
#can be indexed
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout() #This prevents overlap
```

### Figure Size and DPI

```python
fig = plt.figure(figsize = (3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize = (8,2), dpi = 300)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize = (8,4), dpi=300)

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png')

#can also set dpi fig.savefig('my_picture.png',dpi = 300)

```

```python
fig = plt.figure()

# For a legend, labels need to be added
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'Y Squared')

ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Ylabel')

# use location codes (loc = num) or tuple (loc = (x,y)) to specify
ax.legend(loc = 0)
```

```python
fig.savefig('layered_plot.pdf')
```

### Figure Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**3,color = '#2AE416', linewidth = 4, alpha = 0.5)

#color names can be strings ('blue', 'green', etc.), Hex codes
#linewith default is 1 (2 would be double and 0.5 would be half)
ax.plot(x,y,color = '#7B26DC', linewidth = 4, alpha = 0.5,linestyle = '-.'
       ,marker = '3', markersize = 20)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#color names can be strings ('blue', 'green', etc.), Hex codes
#linewith default is 1 (2 would be double and 0.5 would be half)
ax.plot(x,y,color = '#7B26DC', linewidth = 4,linestyle = '-'
       ,marker = 'o', markersize = 20, markerfacecolor='yellow', markeredgewidth = 3, markeredgecolor = 'green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = '#7B26DC', lw = 2, ls = '--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
